# LOGISTIC REGRESSION

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#%matplotlib in line 
```

```python
from sklearn.datasets import make_classification 
```

```python
""""
Create a dataset
n_classes = number of classes
n_features = number of features
n_samples = number of samples
"""

X, Y = make_classification(n_samples=1000, n_features=10, n_classes=2, random_state =42)
```

Convert X to dataframe

```python
pd.DataFrame(X).head()
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2, random_state = 42)
```

### Model Training - Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logisticRegression = LogisticRegression ()
```

```python
# Train Model
logisticRegression.fit(X_train, Y_train)
```

```python
# Predictions
Y_pred = logisticRegression.predict(X_test)
print(Y_pred)
```

```python
# Predicition Probabilites of the classes

#LogisticRegression.predict_proba(X_test)
```

### Performance Metrices

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report 
```

```python
score = accuracy_score(Y_test, Y_pred)
print(f"Accuracy = {score}")
```

```python
_confusion_matrix = confusion_matrix(Y_test, Y_pred)
print(_confusion_matrix)

```

```python
_classification_report = classification_report(Y_test, Y_pred)
print(_classification_report)
```

# Logistic Regression For Multiclass Classification Problem

```python
""""
Create a dataset
n_classes = number of classes
n_features = number of features
n_samples = number of samples
n_informative = number of informative features
"""

X, Y = make_classification(n_informative=3, n_samples=1000, n_features=10, n_classes=3, random_state =42)
```

```python
### Train Test Split

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2, random_state = 42)
```

```python
### Model Training - Logistic Regression
from sklearn.linear_model import LogisticRegression
logisticRegression = LogisticRegression (multi_class='ovr')
```

```python

# Train Model
logisticRegression.fit(X_train, Y_train)
```

```python

# Predictions
Y_pred = logisticRegression.predict(X_test)
print(Y_pred)
```

```python
# Predicition Probabilites of the classes

#LogisticRegression.predict_proba(X_test)

### Performance Metrices
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report 
score = accuracy_score(Y_test, Y_pred)
print(f"Accuracy = {score}")
_confusion_matrix = confusion_matrix(Y_test, Y_pred)
print(_confusion_matrix)


_classification_report = classification_report(Y_test, Y_pred)
print(_classification_report)
```
